# K-Means Clustering Algorithm

K-Means Clustering is an unsupervised learning algorithm that is used to solve the clustering problems in machine learning or data science.

K-Means Clustering is an Unsupervised Learning algorithm, which groups the unlabeled dataset into different clusters. Here K defines the number of pre-defined clusters that need to be created in the process, as if K=2, there will be two clusters, and for K=3, there will be three clusters, and so on.
It is an iterative algorithm that divides the unlabeled dataset into k different clusters in such a way that each dataset belongs only one group that has similar properties.



**Importing Libraries**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
```

```python
data=pd.read_csv('Mall_Customers.csv')
```

```python
data.head()
```

**Feature Extraction**

```python
X = data.iloc[:, [3, 4]].values  #extracting useful features(Annual income and Spending scores)
```

```python
X.shape
```


**Finding the optimal number of clusters using the elbow method**

```python
#Step-2: Finding the optimal number of clusters using the elbow method
from sklearn.cluster import KMeans
wcss_list= []  #Initializing the list for the values of WCSS

#Using for loop for iterations from 1 to 10.
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state= 42)
    kmeans.fit(X)
    wcss_list.append(kmeans.inertia_)

```

```python
plt.plot(np.arange(1, 11), wcss_list,linestyle='dashed',color='g')
plt.title('The Elobw Method Graph')
plt.xlabel('Number of clusters(k)')
plt.ylabel('wcss_list')
plt.show()
```

## Inertia  measures the sum of squared distances between data points and their assigned cluster centroids .

## Ideal No of Clusters is 5

```python
#Step- 3: Training the K-means algorithm on the training dataset
kmeans = KMeans(n_clusters=5, init='k-means++', random_state= 42)
y_predict= kmeans.fit_predict(X)
```

```python
y_predict
```

```python
X[:,0]
```

```python
# visualizing clusters
# Clusters have been classified correctly, below 5 or above 5 clustering will not be proper .
plt.scatter(X[:,0],X[:,1],c=y_predict)
```

```python
#creating cluster column in data
data["Cluster"]=y_predict
```

```python
data.head()
```
