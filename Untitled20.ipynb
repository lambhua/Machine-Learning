```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import sklearn
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
from sklearn.linear_model import LogisticRegression
```

```python
df=pd.read_csv('Telco_Customer_Churn.csv')
df.head()
```

```python
df.describe().transpose()
```

```python
df.info()
```

```python
df.shape  # data has 7043 rows and 21 columns
```

TotalCharges column has empty strings , we will replace empty strings with Null values and than drop those rows


```python
df['TotalCharges']=df['TotalCharges'].replace(' ',np.NaN)
```

```python
df['TotalCharges'].isnull().sum()
```

```python
df.dropna(inplace=True) # dropping rows with null values
```

```python
df['TotalCharges'].isnull().sum()
```

```python
df['TotalCharges'].min(),df['TotalCharges'].max()
```

```python
sns.countplot(data=df,x='InternetService',hue='Churn')
```

# From above figure we see that highest churn has been in Fiber Optics and than in DSL users

```python
sns.countplot(data=df,x='gender',hue='Churn')
```

```python
df.drop('customerID',axis=1,inplace=True)
```

```python
df.replace({"Yes":1, "No":0}, inplace = True)
```

```python
data=pd.get_dummies(df)
```

```python
data.head()
```

```python
lr=LogisticRegression(max_iter=100)
```

```python
x=data.drop('Churn',axis=1)
```

```python
y=data['Churn']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = .25, random_state = 33)
sc=StandardScaler()
sc.fit_transform(x_train)
sc.fit_transform(x_test)
```

```python
lr.fit(x_train,y_train)
```

```python
lr.score(x_train,y_train)
```

```python
lr.score(x_test,y_test)
```

**PREDICTIONS**

```python
y_pred=lr.predict(x_test)
```

```python
confusion_matrix(y_test,y_pred)
```

```python
accuracy_score(y_test,y_pred)
```

```python
classification_report(y_test,y_pred)
```
